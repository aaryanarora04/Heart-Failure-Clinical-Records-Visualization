```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(bins=40, grid=False, figsize=(20,15))
plt.show()
```

```python
#Distribution of age in dataset.
sns.displot(x='age', data=df, bins=40, kde= True)
plt.show()
```

```python
#how many people have diabetes?
df.diabetes.value_counts().plot(kind= 'pie', figsize=(8,6))
plt.title("Diabetes")
plt.legend(["Diabetes", "No Diabetes"])
plt.show()
```

```python
pd.crosstab(df.sex, df.DEATH_EVENT).plot(kind='bar')
plt.title("Count of Male and Female and Death")
plt.xlabel("0= Female, 1= Male")
plt.xticks(rotation = 0)
plt.show()
```

```python
#co-relation between high blood pressure and death
df['high_blood_pressure'].corr(df['DEATH_EVENT'])
```

```python
pd.crosstab(df.high_blood_pressure, df.DEATH_EVENT).plot(kind='bar')
plt.title('High Blood Pressure vs Deaths')
plt.xlabel("0= No High Pressure, 1= High Pressure")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

```python
#co-relation between smoking and death
df['smoking'].corr(df['DEATH_EVENT'])
```

```python
sns.countplot( x= 'smoking', data= df, hue = 'DEATH_EVENT')
plt.title("Smoking vs Death")
plt.xlabel("0= Non Smoker, 1= Smoker")
plt.show()
```

```python
plt.scatter(data=df , x='serum_creatinine', y='serum_sodium')
plt.title('Serum_creatinine vs Serum_sodium')
plt.xlabel('Serum_creatinine')
plt.ylabel('Serum_sodium')
plt.show()
```

```python
sns.displot(x='platelets', data=df, bins=30, kde=True, color='salmon')
plt.show()
```

```python
sns.displot(x='serum_sodium', data=df, bins=20, color='green')
plt.show()
```

```python
df.DEATH_EVENT.value_counts().plot(kind='pie', figsize=(8,6))
plt.title('Death Rate')
plt.legend(['No Death','Death'])
plt.show()
```

```python
plt.scatter(data=df , x='DEATH_EVENT', y='platelets')
plt.title('Count of Platelets vs Death')
plt.xlabel('Death')
plt.ylabel('No. of Platelets')
plt.show()
```
